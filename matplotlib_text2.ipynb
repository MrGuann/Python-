```python
list_year = [
    2006,
    2007,
    2008,
    2009,
    2010,
    2011,
    2012,
    2013,
    2014,
    2015,
    2016,
    2017    
]

list_gdp  = [
    219438.50,
    270232.30,
    319515.50,
    349081.40,
    413030.30,
    489300.60,
    540367.40,
    595244.40,
    643974.00,
    689052.10,
    744127.20,
    827122.00    
]

list_gdp_growth  = [
    12.70,
    14.20, 
    9.70,  
    9.40,  
    10.60, 
    9.50,  
    7.90,  
    7.80,  
    7.30,  
    6.90,  
    6.70,  
    6.90  
]
```

```python
import matplotlib.pyplot as plt
from matplotlib.pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
```

```python
fig = plt.figure(figsize = (12,10))

ax1 = fig.add_subplot(1,1,1)
ax1.bar(list_year, list_gdp, label = 'GDP总量')
ax1.set_xlabel('年份',fontsize = 15)
ax1.set_ylabel('GDP总量/万亿',fontsize = 15)
ax1.legend(loc = 'upper left',fontsize = 20)


ax2 = ax1.twinx()
ax2.plot(list_year, list_gdp_growth, color = 'red', label = 'GDP增长率')
ax2.set_ylabel('GDP增长率/%',fontsize = 15)
ax2.legend(loc = 'upper right',fontsize = 20)
ax2.set_ylim(0,20)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('2006-2017年中国GDP总量/增长率',fontsize = 30)
plt.show()
```

```python
fig = plt.figure(figsize = (12,10))

ax1 = fig.add_subplot(2,1,1)
ax1.bar(list_year, list_gdp, label = 'GDP总量')
ax1.set_xlabel('年份',fontsize = 15)
ax1.set_ylabel('GDP总量/万亿',fontsize = 15)
ax1.legend(loc = 'upper left',fontsize = 20)


ax2 = fig.add_subplot(2,1,2)
ax2.plot(list_year, list_gdp_growth, color = 'red', label = 'GDP增长率')
ax2.set_ylabel('GDP增长率/%',fontsize = 15)
ax2.legend(loc = 'upper right',fontsize = 20)
ax2.set_ylim(0,15)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('2006-2017年中国GDP总量/增长率',fontsize = 30)
plt.tight_layout()
plt.show()
```
